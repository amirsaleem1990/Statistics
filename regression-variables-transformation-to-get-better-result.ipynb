### idea ye h k regression me ham apny numerical vairbales ko transfrom kar k check karty h kya wo targes sy zyada behtar correlated hen ya nahi, yahan me ny sqrt, log2, log10, and polynomial (till degree 4) tak check kar k sab sy zyada jo version correlated h target k sath us variable ko us me convert kar dya h

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("data.csv")
```

```python
num_vars = df.dtypes[(df.dtypes == 'int64') | (df.dtypes == 'float64')].index
df_num = df[num_vars]
y_train = df_num['RentAmount']
df_num.head()
```

```python
df_num.columns
```

```python
vars_to_modifiy = ['','','',''] # specify which variables you need to modify
df_for_modification = df_num[vars_to_modifiy]
most_plausibles = []
for column in df_for_modification.columns:
    orignal = df_for_modification[column]
    sq = np.sqrt(df_for_modification[column])
    lg10 = np.log10(df_for_modification[column]+0.0000001)
    lg2 = np.log2(df_for_modification[column]+0.0000001)
    pol2 = df_for_modification[column] ** 2
    pol3 = df_for_modification[column] ** 3
    pol4 = df_for_modification[column] ** 4
    minn = []
    for i in ['sq', 'lg10', 'lg2', 'pol2', 'pol3', 'pol4', 'orignal']:
        minn.append((i,np.corrcoef(eval(i), y_train).min()))
    most_plausibles.append((column, sorted(dict(minn), key=(lambda key:dict(minn)[key]), reverse=True)[0]))
```

```python
for i in most_plausibles:
    if i [1] ==  'sq':
        df[i[0]] = np.sqrt(df[i[0]])
    elif i[1] == 'lg10':
        df[i[0]] = np.log10(df[i[0]]+ 0.0000001)
    elif i[1] == 'lg2':
        df[i[0]] = np.log2(df[i[0]]+ 0.0000001)
    elif i[1] == 'pol2':
        df[i[0]] = df[i[0]] ** 2
    elif i[1] == 'pol3':
        df[i[0]] = df[i[0]] ** 3
    elif i[1] == 'pol4':
        df[i[0]] = df[i[0]] ** 4
    elif i[1] == 'orignal':
        pass
```
